```python
import pandas as pd
```

```python
# read a remote CSV file
data_url = 'https://raw.githubusercontent.com/alexzalox/NASAinYourNeighborhood/main/MERRA2.avgM_2d_pm25_admin0x.v01.19800101-20221231.csv'
df = pd.read_csv(data_url, header=13)
df.head(5)
```

```python
# Observe the dtypes
df.dtypes
```

```python
# Convert date string to datetime object
df['date'] = pd.to_datetime(df['date']) + pd.offsets.MonthEnd()

# Create a new column:"days_in_month" to recode the total days of the month
df['days_in_month'] = df.apply(lambda row: pd.Timestamp(year=row['date'].year, month=row['date'].month, day=1).days_in_month, axis=1)

df.head(5)
```

```python
# check the dtypes again
df.dtypes
```

```python
# set day counts of the month
df_month_total = df.iloc[:, 1:-1].multiply(df['days_in_month'], axis=0)
df_month_total.head(5)
```

```python
# Create cumulative df
df_cumulated = df_month_total.cumsum()
df_cumulated['date'] = df['date']
df_cumulated.set_index('date', inplace=True)
df_cumulated.head(12)
```

```python
# Remove the underscores in the country names
df_cumulated.columns = [col.replace('_', ' ') for col in df_cumulated.columns]
```

```python
# write a df_year_cumulated.csv (for teammate used, this cell is no correlation with this notebook.)

# copy index from df to df_month_total
df_month_total['date'] = df['date']
df_month_total.set_index('date', inplace=True)

# # use resample('Y') to build a year-cumulated df
df_year_cumulated = df_month_total.resample('Y').sum().cumsum()
df_year_cumulated
# write a df_year_cumulated.csv
# df_year_cumulated.to_csv('./pm25_year_cumulated.csv', sep=',', index=True)
```

```python
# Remove countries that never ranked in the top 20
country_reserved = set()
for index, row in df_cumulated.iterrows():
    country_reserved |= set(row[row > 0].sort_values(ascending=False).head(20).index)
df_cumulated = df_cumulated[list(country_reserved)]
```

```python
# Observe the shape, now we only have 28 countries left
df_cumulated.shape
```
